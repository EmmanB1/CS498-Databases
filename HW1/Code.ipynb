```python
import pandas as pd
df = pd.read_csv("/content/YouTubeDataset_withChannelElapsed.csv")
df
```

```python
#aggregate the data for each unique videoId
agg_df = df.groupby("videoId").agg('sum')
agg_df
```

```python
# 1. Calculate the average number of likes per video.
agg_df['videoLikeCount'].mean()
```

```python
agg_df.columns
```

```python
#2. Compute the engagement rate for each video
#likes + dislikes + comments / totalviews
agg_df['engagement_rate'] = (agg_df['videoLikeCount'] + agg_df['VideoCommentCount'] + agg_df['videoDislikeCount']) / agg_df['videoViewCount']
agg_df['engagement_num'] = agg_df['videoLikeCount'] + agg_df['VideoCommentCount'] + agg_df['videoDislikeCount']
display(agg_df['engagement_rate'])
print(agg_df['engagement_rate'].agg('mean'))
```

```python
# 3. Calculate the combined storage size for a video in megabytes (MB)
# Video = 500 bytes metadata, Thumbnail = 200 KB (200,000)
# We know the metadata and thumbnail... what is the storage size for a specific video? What about the size of the contents of the video?
# I'm confused about if there's more that I should be seeing
# 200,500 B
#.200500 MB
STORAGE_SIZE = .2005
```

```python
# 4. Estimate the total data transfer required for all engagements with videos in the dataset
tot_data_trans = agg_df["engagement_num"].sum() * STORAGE_SIZE
tot_data_trans
```

```python
# 5. Estimate the number of views per minute, assuming the dataset spans 5 years with views uniformly distributed over time.
# years, days/year, hours/day, min/hour
view_per_min = agg_df["videoViewCount"].sum() / (5 * 365 * 24 * 60)
view_per_min
```

```python
# 6. Estimate the amount of data accessed per minute to display video details to viewers.
data_per_min = tot_data_trans / (5 * 365 * 24 * 60)
data_per_min
```
